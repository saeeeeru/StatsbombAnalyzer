# 準備

```python
import os, requests

import statsbomb as sb
import pandas as pd
```

```python
BASE_URL = 'https://raw.githubusercontent.com/statsbomb/open-data/master/data'
```

# 大会の確認

```python
comps = sb.Competitions()
print(len(comps))
```

```python
comps_df = comps.get_dataframe()
```

```python
comps_df[comps_df.competition_name.isin(['Champions League', 'Premier League'])]
```

# CL

## 試合

```python
competition_id = 16
matches_df = pd.concat([pd.DataFrame(requests.get(f'{BASE_URL}/matches/{competition_id}/{season_id}.json').json()) for season_id in comps_df[comps_df.competition_id==competition_id].season_id.tolist()]).reset_index()
```

```python
c_list = ['competition', 'season', 'home_team', 'away_team']
for c in c_list:
    k_list = [k for k in list(matches_df.at[0, c].keys()) if k.endswith('name')]
    for k in k_list:
        matches_df[k] = matches_df[c].apply(lambda x: x[k])
```

```python
matches_df[[c for c in matches_df.columns if c.endswith('name') or c.endswith('id')]]
```

## イベントデータを抜いてみる

```python
match_id = 18245
events_df = pd.DataFrame(requests.get(f'{BASE_URL}/events/{match_id}.json').json()).dropna(subset=['player']).reset_index(drop=True)
```

```python
events_df.columns
```

```python
events_df.head()
```

```python
c_list = ['type', 'possession_team', 'play_pattern', 'player']
for c in c_list:
    k_list = [k for k in list(events_df.at[0, c].keys()) if k.endswith('name')]
    for k in k_list:
        events_df[f'{c}_{k}'] = events_df[c].apply(lambda x: x[k] if x else None)
```

```python
events_df[[c for c in events_df.columns if c.endswith('name')]].head(20)
```
